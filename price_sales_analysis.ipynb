```python
## 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
## 2. Load Dataset
data = pd.read_csv('data/mobile_price_classification.csv')  # Adjust the path as needed
```

```python
# Settings for visualizations
%matplotlib inline
sns.set(style="whitegrid")
```

```python
# Display basic dataset information
print("Dataset Shape:", data.shape)
print("Dataset Info:")
data.info()
print("\nSample Data:\n", data.head())
```

```python
## 3. Data Cleaning
# Check for duplicates
data.drop_duplicates(inplace=True)
```

```python
# Check for missing values
print("Missing Values:\n", data.isnull().sum())
```

```python
## 4. Exploratory Data Analysis (EDA)
# Price Range Distribution
plt.figure(figsize=(8, 5))
sns.countplot(x='price_range', data=data, palette='coolwarm')
plt.title("Distribution of Mobile Phones by Price Range")
plt.xlabel("Price Range")
plt.ylabel("Frequency")
plt.show()
```

```python
# Battery Power Distribution by Price Range
plt.figure(figsize=(10, 6))
sns.boxplot(x='price_range', y='battery_power', data=data, palette='viridis')
plt.title("Battery Power Distribution Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Battery Power")
plt.show()
```

```python
# RAM Distribution by Price Range
plt.figure(figsize=(10, 6))
sns.boxplot(x='price_range', y='ram', data=data, palette='Set2')
plt.title("RAM Distribution Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("RAM (MB)")
plt.show()
```

```python
# Internal Memory Distribution by Price Range
plt.figure(figsize=(10, 6))
sns.boxplot(x='price_range', y='int_memory', data=data, palette='pastel')
plt.title("Internal Memory Distribution Across Price Ranges")
plt.xlabel("Price Range")
plt.ylabel("Internal Memory (GB)")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(12, 10))
correlation = data.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap of Mobile Phone Features")
plt.show()
```

```python
# Insights into battery and RAM as they relate to price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='ram', y='battery_power', hue='price_range', data=data, palette='Set1')
plt.title("Relationship Between RAM, Battery Power, and Price Range")
plt.xlabel("RAM (MB)")
plt.ylabel("Battery Power")
plt.show()
```

# Mobile Price Sales Feature Analysis

This analysis examines the relationship between mobile phone features and sales prices. Various visualizations are used to provide insights into how specifications like battery power, RAM, and internal memory affect a phone's price range.

## Summary of Findings

### 1. Price Range Distribution
- **Description**: Shows the number of mobile phones across different price categories.
- **Insight**: The distribution allows us to see how many phones are in each price range, highlighting the balance or concentration of price categories in the dataset.

### 2. Battery Power Distribution by Price Range
- **Description**: A box plot comparing battery power across price ranges.
- **Insight**: Higher-priced phones tend to have higher battery power, showing that larger batteries are often a feature in premium phones.

### 3. RAM Distribution by Price Range
- **Description**: A box plot illustrating how RAM varies with price range.
- **Insight**: RAM capacity generally increases with price. Higher-priced phones tend to have more RAM, supporting better performance, especially for multitasking and running applications.

### 4. Internal Memory Distribution by Price Range
- **Description**: A box plot displaying the distribution of internal memory across price ranges.
- **Insight**: Internal memory, or storage, is usually higher in more expensive phones. This feature is important for users who store media and applications, and it is more prevalent in higher-priced models.

### 5. Correlation Heatmap
- **Description**: A heatmap showing correlations between features like RAM, battery power, and price range.
- **Insight**: Strong correlations are observed between price range and features like RAM and battery power. This implies that these specifications significantly impact the phone's pricing.

### 6. RAM and Battery Power Relationship by Price Range
- **Description**: A scatter plot showing the relationship between RAM and battery power, categorized by price range.
- **Insight**: Higher-priced phones tend to have both higher RAM and battery power, suggesting that these are important features for premium models.

## Conclusion
This analysis highlights the influence of key mobile phone specifications on pricing. Battery power, RAM, and internal memory emerge as significant factors that differentiate mobile phones across price ranges, offering insights into feature importance for customers.
